```python
#names
names = ["Alex", "Holly", "Steve", "Chris", "Lucheins", "Norman", "Karolina", "Anyssa", "Rob", "Ryan", "Elena", "Andy", "Kirsty", "Joseph", "Matt", "Julia"]

#willingness
will = [85, 35, 100, 92, 53, 46, 11, 73, 25, 5, 95, 98, 60, 95, 98, 90]

#talent
talent = [85, 65, 55, 92, 87, 79, 61, 69, 41, 40, 70, 60, 50, 60, 88, 85]

#enthusiasm
enthu = [80, 60, 100, 100, 39, 55, 28, 67, 39, 37, 95, 70, 40, 100, 92, 90]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
from scipy import stats
```

```python
plt.xlabel("Willingness to Perform (%)", fontsize=16)
plt.ylabel("Talent when Performing (%)", fontsize=16)
plt.title("MicroBooNE at Karaoke", fontsize=16)

plt.scatter(will, talent, c=enthu, cmap='viridis', marker='*')
#plt.colorbar(pad=-2)
plt.colorbar().set_label('Enthusiasm when Performing (%)', rotation=270, fontsize=14)
plt.plot(np.unique(will), np.poly1d(np.polyfit(will, talent, 1))(np.unique(will)), ":", color="purple")

plt.plot()
```
