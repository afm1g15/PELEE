# Look for overlap in Nue and NuMu Datasets

```python
import pandas as pd
```

```python
#First, load in the csv files with the data

#nue
nue_points = pd.read_csv("nue_full.csv")

#numu
numu_points = pd.read_csv("numu_full.csv")
```

```python
#plot the types shown? How to find overlap?
#look at entry numbers?
```

```python
print(nue_points.entry)
```

```python
print(numu_points.entry)
```

```python
#see if there's any matches
matches = 0
match_list = []
pos_list = []

for i in numu_points.entry:
    count = 0
    for j in nue_points.entry:
        if i == j:
            matches = matches + 1
            match_list.append(i)
            pos_list.append(count)
            count = count + 1 
        else:
            count = count + 1   
            
print("number of matches = ", matches)
print("match numbers = ", match_list)
print("pos numbers = ", pos_list)
```
